```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib
```

```python
data = {
    "age": [25, 45, 35, 23, 52, 40, 29, 60],
    "claim_amount": [5000, 20000, 7000, 3000, 50000, 15000, 4000, 60000],
    "previous_claims": [0, 3, 1, 0, 5, 2, 0, 6],
    "fraud": [0, 1, 0, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
df.head()
```

```python
X = df[["age", "claim_amount", "previous_claims"]]
y = df["fraud"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model Accuracy:", model.score(X_test, y_test))
```

```python
import joblib

joblib.dump(model, r"C:\xampp\htdocs\mini project\backend\fraud_model.pkl")
```
